### Загрузка и первичный обзор

```python
#загрузить данные через pandas

import pandas as pd
df = pd.read_csv('google_books_dataset.csv')
```

```python
#вывести первые/последние строки

df.head()
```

```python
#вывести первые/последние строки

df.tail()
```

```python
#посмотреть структуру данных

df.info()
```

```python
#проверить типы данных и при необходимости привести их к корректным типам

#столбец published_date в формат даты и времени
df['published_date'] = pd.to_datetime(df['published_date'], errors='coerce')
#столбец isbn_13 — в строковый тип, поскольку он представляет собой идентификатор
df['isbn_13'] = df['isbn_13'].astype(str).replace('\.0$', '', regex=True).replace('nan', pd.NA)
df.info()
```

### Обработка данных

Пропущенные значения

```python
df.isnull()
```

```python
df.isna()
```

```python
missing_values = df.isnull().sum()
print(missing_values)
```

```python
threshold = 0.8 * len(df)
columns_to_drop = missing_values[missing_values > threshold].index.tolist()

print(f"DF shape: {df.shape}")
print(f"Columns to drop: {columns_to_drop}")

df_cleaned = df.drop(columns=columns_to_drop)
print(f"Cleaned DF shape: {df_cleaned.shape}")

missing_values_after_drop = df_cleaned.isnull().sum()
print("Missing values after dropping columns:")
print(missing_values_after_drop[missing_values_after_drop > 0])
```

Дубликаты

```python
duplicate_rows = df_cleaned[df_cleaned.duplicated()]
print(f"Number of duplicate rows: {len(duplicate_rows)}")
duplicate_rows.head()
```

```python
#описательная статистика
df_cleaned.describe()
```

```python
for column in df_cleaned.columns:
    if column in ['book_id', 'preview_link', 'info_link']:
        print(f"Column: {column}")
        print(f"Number of unique values: {df_cleaned[column].nunique()}\n")

    print(f"Column: {column}")
    print(f"Number of unique values: {df_cleaned[column].nunique()}")
    print(f"Top 10 value counts:\n{df_cleaned[column].value_counts().head(5)}\n")
```

```python
correlation_matrix = df_cleaned.corr(numeric_only=True)
print(correlation_matrix)
```

### Визуализация данных

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
sns.histplot(df_cleaned['page_count'].dropna(), kde=True)
plt.title('Distribution of Page Count')
plt.xlabel('Page Count')
plt.ylabel('Frequency')

plt.subplot(1, 2, 2)
sns.histplot(df_cleaned['ratings_count'].dropna(), kde=True)
plt.title('Distribution of Ratings Count')
plt.xlabel('Ratings Count')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

```python
df_cleaned['publication_year'] = df_cleaned['published_date'].dt.year
publication_trends = df_cleaned['publication_year'].value_counts().sort_index()

plt.figure(figsize=(12, 6))
sns.lineplot(x=publication_trends.index, y=publication_trends.values)
plt.title('Publication Trends Over Years')
plt.xlabel('Year')
plt.ylabel('Number of Publications')
plt.grid(True)
plt.show()
print("Line chart for publication trends generated.")
```

```python
plt.figure(figsize=(12, 6))
top_authors = df_cleaned['authors'].value_counts().head(10)
sns.barplot(x=top_authors.index, y=top_authors.values, palette='viridis')
plt.title('Top 10 Authors by Number of Publications')
plt.xlabel('Author')
plt.ylabel('Number of Publications')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
print("Bar chart for top 10 authors generated.")
```

```python
plt.figure(figsize=(15, 6))

plt.subplot(1, 2, 1)
top_categories = df_cleaned['categories'].value_counts().head(10)
sns.barplot(x=top_categories.index, y=top_categories.values, palette='viridis', hue=top_categories.index, legend=False)
plt.title('Top 10 Categories by Number of Publications')
plt.xlabel('Category')
plt.ylabel('Number of Publications')
plt.xticks(rotation=45, ha='right')

plt.subplot(1, 2, 2)
top_languages = df_cleaned['language'].value_counts().head(10)
sns.barplot(x=top_languages.index, y=top_languages.values, palette='magma', hue=top_languages.index, legend=False)
plt.title('Top 10 Languages by Number of Publications')
plt.xlabel('Language')
plt.ylabel('Number of Publications')
plt.xticks(rotation=45, ha='right')

plt.tight_layout()
plt.show()
print("Bar charts for top 10 categories and top 10 languages generated.")
```

```python
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
sns.boxplot(x='buyable', y='page_count', data=df_cleaned, palette='viridis')
plt.title('Page Count Distribution by Buyable Status')
plt.xlabel('Buyable Status')
plt.ylabel('Page Count')

plt.subplot(1, 2, 2)
sns.boxplot(x='buyable', y='ratings_count', data=df_cleaned, palette='magma')
plt.title('Ratings Count Distribution by Buyable Status')
plt.xlabel('Buyable Status')
plt.ylabel('Ratings Count')

plt.tight_layout()
plt.show()
```

### Группировки и агрегации

```python
category_stats = df_cleaned.groupby('categories').agg(
    mean_page_count=('page_count', 'mean'),
    median_page_count=('page_count', 'median'),
    mean_ratings_count=('ratings_count', 'mean'),
    median_ratings_count=('ratings_count', 'median')
).reset_index()

category_stats.head()
```

```python
author_publication_counts = df_cleaned.groupby('authors').size().reset_index(name='publication_count')
author_publication_counts.head()
```

```python
language_page_counts = df_cleaned.groupby('language')['page_count'].sum().reset_index()
language_page_counts.head()
```

```python
pivot_table_publisher_language = df_cleaned.pivot_table(
    index='publisher',
    columns='language',
    values='page_count',
    aggfunc='mean'
)

pivot_table_publisher_language.head()
```
